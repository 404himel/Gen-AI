```python
from transformers import pipeline
Classifier = pipeline('sentiment-analysis')
result = Classifier('Hi, This is XYZ')
print(result)
```

```python
classifier_2 = pipeline(task='sentiment-analysis', model='rayschwartz/text-classification')
result = classifier_2(' He is a good boy.This product is good.')
print(result)
```

```python
qa_model = pipeline('question-answering')
question = """How would you optimize a Hugging Face Transformer model for faster inference in production — especially when deploying on limited cloud resources or through an API?"""
context = """You’re working on a project that involves deploying a Hugging Face Transformer model (for example, a text classification or summarization model) into a production workflow.
The team wants to use the model within an N8N automation or a web application, possibly through the Hugging Face Inference API or by hosting the model locally.
You’ll need to ensure good performance, scalability, and security."""
result = qa_model(question = question, context = context)
print(result)
```
